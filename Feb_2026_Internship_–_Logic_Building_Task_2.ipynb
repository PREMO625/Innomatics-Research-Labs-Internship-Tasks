## Problem 1: Unique Words in a Sentence

```python
# Input sentence
sentence = input("Enter sentence: ")

# Split the sentence into words and convert to lowercase for case-insensitive comparison
words = sentence.lower().split()

# Convert the list of words to a set to get unique words
unique_words = set(words)

# Display the count of unique words
print(f"Unique words count: {len(unique_words)}")

# Display the unique words themselves
print(f"Unique words: {unique_words}")
```

## Problem 2: Highest Salary from Employee Data

```python
# Example Input
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000,
    "Priya": 80000
}

# Initialize variables to keep track of the highest salary and the corresponding employee
highest_salary = 0
highest_paid_employee = ""

# Iterate through the dictionary to find the employee with the highest salary
for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_paid_employee = name

# Display the result
print(f"Highest Salary: {highest_paid_employee} - {highest_salary}")
```

## Problem 3: Find Maximum and Minimum Values

```python
# Example Input
numbers = [45, 22, 89, 10, 66, 5, 95, 33]

# Initialize maximum and minimum with the first element of the list
if not numbers:
    max_val = None
    min_val = None
else:
    max_val = numbers[0]
    min_val = numbers[0]

    # Iterate through the list to find the maximum and minimum values
    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

# Display the results
print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

## Problem 4: Count Products Above a Price Threshold

```python
# Example Input
prices = [450, 1200, 899, 1500, 300]

# Initialize a counter for products above the price threshold
count_above_threshold = 0

# Iterate through the list of prices
for price in prices:
    # Check if the current price is greater than 1000
    if price > 1000:
        count_above_threshold += 1

# Display the result
print(f"Products above 1000: {count_above_threshold}")
```

## Problem 5: Calculate Attendance Percentage

```python
# Example Input
attendance = ["P", "P", "A", "P", "P"]

# Count the number of 'P' (Present) entries
present_count = attendance.count("P")

# Calculate the total number of days
total_days = len(attendance)

# Calculate attendance percentage, handling division by zero if attendance list is empty
if total_days > 0:
    attendance_percentage = (present_count / total_days) * 100
else:
    attendance_percentage = 0.0 # Or handle as an error, depending on requirements

# Display the result
print(f"Attendance Percentage: {attendance_percentage}")
```

## Problem 7: Count Character Frequency

```python
# Example Input
input_string = "python"

# Initialize an empty dictionary to store character counts
char_frequency = {}

# Iterate through each character in the string
for char in input_string:
    # Increment the count for the character, or initialize to 1 if not seen before
    char_frequency[char] = char_frequency.get(char, 0) + 1

# Display the character frequency
print(char_frequency)
```

## Problem 10: Calculate Average Marks

```python
# Example Input
marks = [85, 70, 60, 75]

# Calculate the sum of marks
total_marks = sum(marks)

# Calculate the number of subjects (or marks)
num_subjects = len(marks)

# Calculate the average, handling division by zero if the list is empty
if num_subjects > 0:
    average_marks = total_marks / num_subjects
else:
    average_marks = 0.0 # Or handle as an error

# Display the average marks
print(f"Average Marks: {average_marks}")
```
