### 1. User Login Check

**Problem Statement:** Given a username and password, check whether login is successful.

**Requirements:**
* Print "Login Successful" if both username and password match.
* Otherwise print "Invalid Credentials".

**Real-World Application:** Authentication systems.

```python
username = "admin"
password = "1234"

# Simulate user input or predefined values
input_username = "admin"
input_password = "1234"

if input_username == username and input_password == password:
    print("Login Successful")
elif input_username == username and input_password != password:
    print("Invalid Credentials - Incorrect Password")
else:
    print("Invalid Credentials - Incorrect Username")
```

### 2. Pass / Fail Analyzer

**Problem Statement:** Analyze student results.

**Requirements:**
* A student passes if marks â‰¥ 50.
* Count the total number of pass students.
* Count the total number of fail students.
* Print the final result clearly.

**Real-World Application:** Academic evaluation systems.

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total students: {len(marks)}")
print(f"Pass students: {pass_count}")
print(f"Fail students: {fail_count}")
```

### 3. Simple Data Cleaner

**Problem Statement:** Clean and standardize user names.

**Requirements:**
* Remove extra spaces.
* Convert all names to lowercase.
* Print the cleaned list.

**Real-World Application:** Data preprocessing before analysis.

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []
for name in names:
    cleaned_names.append(name.strip().lower())

print("Original names:", names)
print("Cleaned names:", cleaned_names)
```

### 4. Message Length Analyzer

**Problem Statement:** Analyze message sizes.

**Requirements:**
* Print the length of each message.
* Flag messages longer than 10 characters.

**Real-World Application:** Text filtering and validation systems.

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    msg_length = len(message)
    print(f"Message: '{message}' - Length: {msg_length}", end='')
    if msg_length > 10:
        print(" (FLAG: Longer than 10 characters)")
    else:
        print()
```

### 5. Error Message Detector

**Problem Statement:** Detect error messages from system logs.

**Requirements:**
* Count the number of "ERROR" entries.
* Print the total error count.

**Real-World Application:** Monitoring and log analysis systems.

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for entry in logs:
    if entry == "ERROR":
        error_count += 1

print(f"Total 'ERROR' entries found: {error_count}")
```
